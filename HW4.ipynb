```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklift.metrics import uplift_at_k
from sklift.viz import plot_uplift_preds
from sklift.models import SoloModel, ClassTransformation, TwoModels
from catboost import CatBoostClassifier

from causalml.inference.tree import UpliftTreeClassifier
from causalml.inference.tree import uplift_tree_plot
from IPython.display import Image

%matplotlib inline
pd.options.display.max_columns = 100
```

Загрузим данные из файла

```python
df = pd.read_csv('/home/maxim/Документы/GeekBrains/4 четверть. Машинное обучение в бизнесе/Manuals/lection4_materials/HW_data.csv', sep=',')
```

Поменяем названия колонок 'conversion' -> 'target' и 'offer' -> 'treatment'. Заменим по условию задачи 'No offer' на 0, а остальные значения на 1.

```python
df.rename(columns={'conversion': 'target', 'offer': 'treatment'}, inplace=True)
df.loc[df['treatment'] == 'No Offer', ['treatment']] = 0
df.loc[(df['treatment'] == 'Discount') | (df['treatment'] == 'Buy One Get One'), ['treatment']] = 1
```

Добавим две новые фичи:
- Одна будет принимать значение 0, если последняя покупка была совершена менее 3 месяцев назад и значение 1 если более 4.
- Вторая будет делить значения сумм покупок по квантилям: до 0.2, после 0.5 и между ними.

```python
df['recency_bin'] = 0
df.loc[(df['recency'] > 3), 'recency_bin'] = 1
```

```python
df['history_cat'] = 0
df.loc[(df['history'].between(df['history'].quantile(0.2), df['history'].quantile(0.5))), 'history_cat'] = 1
df.loc[(df['history'].between(df['history'].quantile(0.5), df['history'].quantile(1))), 'history_cat'] = 2
```

Приведем типы данных к int и str. Разобьем данные на тестовую и тренировочную выборки.

```python
df['treatment'] = df['treatment'].astype('int')
df[['zip_code', 'channel']] = df[['zip_code', 'channel']].astype('str')

X_train, X_test, y_train, y_test = train_test_split(
                                                    df.drop(columns=['target']),
                                                    df['target'],
                                                    test_size=0.3,
                                                    random_state=10,
                                                   )
```

Объявим категориальные переменные, словарь с результатами.

```python
cat_features = ['used_discount', 'used_bogo', 'zip_code', 'is_referral', 'channel']

results = {
    'approach': [],
    'uplift@10%': [],
    'uplift@20%': [],
}
```

Создадим и обучим наши модели.

```python
# Solomodel
sm = SoloModel(CatBoostClassifier(iterations=50, thread_count=2, random_state=42, silent=True))
sm = sm.fit(X_train.drop(columns='treatment'),
            y_train,
            X_train['treatment'],
            estimator_fit_params={'cat_features': cat_features})

uplift_sm = sm.predict(X_test)

# ClassTransformation
ct = ClassTransformation(CatBoostClassifier(iterations=50, thread_count=2, random_state=42, silent=True))
ct = ct.fit(X_train.drop(columns='treatment'),
            y_train,
            X_train['treatment'],
            estimator_fit_params={'cat_features': cat_features})

uplift_ct = ct.predict(X_test)

#TwoModels
tm = TwoModels(estimator_trmnt=CatBoostClassifier(iterations=50, thread_count=2, random_state=42, silent=True), 
               estimator_ctrl=CatBoostClassifier(iterations=50, thread_count=2, random_state=42, silent=True), 
               method='vanilla')
tm = tm.fit(X_train.drop(columns='treatment'), y_train, X_train['treatment'],
            estimator_trmnt_fit_params={'cat_features': cat_features}, 
            estimator_ctrl_fit_params={'cat_features': cat_features})

uplift_tm = tm.predict(X_test)
```

Занесем результаты в словарь и выведем сводную таблицу.

```python
results['approach'].extend([
                        'SoloModel', 
                        'ClassTransformation', 
                        'TwoModels',
                        ])

results['uplift@10%'].extend([
    uplift_at_k(y_true=y_test, uplift=uplift_sm, treatment=X_test.treatment, strategy='by_group', k=0.1),
    uplift_at_k(y_true=y_test, uplift=uplift_ct, treatment=X_test.treatment, strategy='by_group', k=0.1),
    uplift_at_k(y_true=y_test, uplift=uplift_tm, treatment=X_test.treatment, strategy='by_group', k=0.1),
])
results['uplift@20%'].extend([
    uplift_at_k(y_true=y_test, uplift=uplift_sm, treatment=X_test.treatment, strategy='by_group', k=0.2),
    uplift_at_k(y_true=y_test, uplift=uplift_ct, treatment=X_test.treatment, strategy='by_group', k=0.2),
    uplift_at_k(y_true=y_test, uplift=uplift_tm, treatment=X_test.treatment, strategy='by_group', k=0.2),
])
```

```python
pd.DataFrame(results)
```

Самый лучший результат показала модель SoloModel.

```python
X_train_tree = pd.get_dummies(X_train.drop(columns='treatment'))
X_test_tree = pd.get_dummies(X_test.drop(columns='treatment'))
```

```python
uplift_model = UpliftTreeClassifier(max_depth=8, min_samples_leaf=200, min_samples_treatment=50,
                                    n_reg=100, evaluationFunction='KL', control_name='control')

uplift_model.fit(X_train_tree.values,
                 treatment=X_train.treatment.map({1: 'treatment1', 0: 'control'}).values,
                 y=y_train)

features = [col for col in X_train_tree]
graph = uplift_tree_plot(uplift_model.fitted_uplift_tree, features)
img = Image(graph.create_png())
img
```
